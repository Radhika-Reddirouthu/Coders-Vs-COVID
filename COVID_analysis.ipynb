### Adding no.of beds, no.of testing labs, Rural population in each state to df9

```python
df9["beds"]=0
df9["labs"]=0
df9["Rural Population"]=0
```

Adding information about no.of hospital beds available in each state from df2 

```python
for i in range(len(df9)):
    for j in range(len(df2)):
        if df9["State/ Union Territory"][i]==df2["State/UT"][j]:
            df9["beds"][i]=df2["NumPublicBeds_HMIS"][j]
            break
        if df9["State/ Union Territory"][i]=="Dadra and Nagar Haveli and Daman and Diu" and (df2["State/UT"][j]=="Dadra & Nagar Haveli" or df2["State/UT"][j]=="Daman & Diu"):
            df9["beds"][i]+=df2["NumPublicBeds_HMIS"][j]
```

Adding information of no.of testing labs available in each state from df4

```python
for i in range(len(df9)):
    for j in range(len(df4)):
        if df9["State/ Union Territory"][i]==df4["state"][j]:
            df9["labs"][i]+=1
        if df9["State/ Union Territory"][i]=="Andaman & Nicobar Islands" and df4["state"][j]=="Andaman and Nicobar Islands":
            df9["labs"][i]+=1
        if df9["State/ Union Territory"][i]=="Jammu & Kashmir" and df4["state"][j]=="Jammu and Kashmir":
            df9["labs"][i]+=1
```

Adding information of rural population from df6

```python
for i in range(len(df9)):
    for j in range(len(df6)):
        if df9["State/ Union Territory"][i]==df6["State / Union Territory"][j]:
            df9["Rural Population"][i]=df6["Rural population"][j]
            break
        if df9["State/ Union Territory"][i]=="Telangana" and df6["State / Union Territory"][j]=="Telengana":
            df9["Rural Population"][i]=df6["Rural population"][j]
            break
        if df9["State/ Union Territory"][i]=="Jammu & Kashmir" and df6["State / Union Territory"][j]=="Jammu and Kashmir":
            df9["Rural Population"][i]=df6["Rural population"][j]
            break
        if df9["State/ Union Territory"][i]=="Andaman & Nicobar Islands" and df6["State / Union Territory"][j]=="Andaman and Nicobar Islands":
            df9["Rural Population"][i]=df6["Rural population"][j]
            break
            
```

```python
x= df9["State/ Union Territory"]
y=np.log(df9["Total Cases"])
z=np.log(df9["Rural Population"])
#w= df8["Gender Ratio"]%10
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.plot(x, y, 'o-',label="Total cases")
ax.plot(x, z, 'o-',label="Rural Population")
ax.legend(loc=2, bbox_to_anchor=(0.6,0.5))
ax.scatter((0.6),(0.5), s=81, c="limegreen", transform=ax.transAxes)
plt.setp(ax.get_xticklabels(), rotation=90, horizontalalignment='right')
plt.title("Rural population vs No.of Cases")
plt.show()
```

From the above graph it is clear that in most of the states, cases are directly proportional to Rural Population. Lock down restrictions and social distancing should be followed properly in rural areas.

```python
x= df9["State/ Union Territory"]
y=np.log(df9["Total Cases"])
z=np.log(df9["beds"])
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.plot(x, y, 'o-',label="Total Cases")
ax.plot(x, z, 'o-',label="beds")
ax.legend(loc=2, bbox_to_anchor=(0.6,0.5))
ax.scatter((0.6),(0.5), s=81, c="limegreen", transform=ax.transAxes)
plt.setp(ax.get_xticklabels(), rotation=90, horizontalalignment='right')
plt.title("No.of Cases vs No.of Beds")
plt.show()
```

From the above graph we can conclude that every state is having proper health care facilities and few states are even ready to face if there is a increase in no.of cases

### Creating a new data frame df10 with date wise cases information

```python
dates=pd.unique(df7['Date'])
dates =pd.to_datetime(dates)
dates=dates.sort_values()
```

```python
dates
```

```python
df10=pd.DataFrame(dates.copy())
df10.columns=['date']
df10["cases"]=0
```

```python
for i in range(len(df10)):
    temp=0
    for j in range(len(df7)):
        if df10["date"][i]==df7["Date"][j]:
            temp+=df7["Positive"][j]
    df10["cases"][i]=temp
        
df10.drop([0,1,29,30,31],inplace=True)
```

```python
df10.head()
```

```python
x= df10["date"]
y=df10["cases"]
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.plot(x, y, 'o-')
plt.setp(ax.get_xticklabels(), rotation=90, horizontalalignment='right')
plt.title("date wise total no.of cases")
plt.show()
```
